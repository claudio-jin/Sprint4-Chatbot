```python
covid = pd.read_excel('dados-teste-covidistress.xlsx')
```

```python
covid.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
#variavel fig recebe as imagens
#axes recebe um conjunto de arrays
#fig-size lado esquerdo, é o width
#fig-size lado direito, é o height


print(type(axes))
fig.suptitle('Todos os Dados')
#imagem recebe o nome "todos os dados"

sns.histplot(ax=axes[0],
#axes indica a posição que o grafico vai estar
#cahma o seaborn, cria um histo

            data=covid['PSS10_avg'],
#na variavel data, vai receber de dados a coluna comprimento abdomen

            kde=True,
#linha que indica a variancia
             
            bins=10,
#quantidade de barras que aparecem

            linewidth=0.7,
#largura da linha
            )

sns.histplot(ax=axes[1],
            data=covid['SLON3_avg'],
            kde=True,
            bins=10,
            linewidth=0.7,
            )

axes[0].set(xlabel='PSS10_avg', ylabel='Contagem')
axes[1].set(xlabel='SLON3_avg', ylabel='Contagem')
plt.subplots_adjust(wspace = 0.35)
```
