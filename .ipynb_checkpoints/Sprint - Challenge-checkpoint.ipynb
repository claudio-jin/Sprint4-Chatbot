<h1>GRÁFICOS REFERENTES AOS DADOS DO ARTIGO COVIDiSTRESS</h1>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

covid = pd.read_excel('dados-teste-covidistress.xlsx')
```

Na célula acima, importamos as bibliotecas pandas, seaborn e matplotlib;

Para um processamento mais rápido dos dados, algumas colunas foram separadas e adicionadas em um arquivo excel.

-Legendas:</br>
    variavel fig recebe as imagens</br>
    axes recebe um conjunto de arrays</br>
    fig-size lado esquerdo, é o width</br>
    fig-size lado direito, é o height</br>
    imagem recebe o nome pelo metodo suptitle</br>
    axes indica a posição que o gráfico vai estar</br>
    sns.histplot = chama o seaborn, cria um histograma</br>
    a variável data, vai receber de dados a coluna desejada</br>
    'kde' = linha que indica a variancia </br>
    bins = quantidade de barras que aparecem</br>
    linwidth = largura da linha</br>


```python
fig, axes = plt.subplots(1, figsize = (10,7))

fig.suptitle('Dados da Coluna PSS10_avg')
sns.histplot(
            data = covid['PSS10_avg'],
            kde=True,
            bins = 10,
            )
axes.set(xlabel='PSS10_avg', ylabel='QTD.população')
```

O gráfico acima representa a média resultante da pesquisa COVIDiSTRESS em relação à escala PSS10

```python
fig, axes = plt.subplots(1, figsize = (10,7))

fig.suptitle('Dados da Coluna SLON3_avg')
sns.histplot(
            data = covid['SLON3_avg'],
            kde=True,
            bins = 10,
            color = 'red'
            )
axes.set(xlabel='SLON3_avg', ylabel='QTD.população')
```

O gráfico acima representa a média resultante da pesquisa COVIDiSTRESS em relação à escala SLON3

```python
fig, axes = plt.subplots(1, figsize = (10,7))

fig.suptitle('Dados da Coluna SPS_avg')
sns.histplot(
            data = covid['SPS_avg'],
            kde=True,
            bins = 10,
            color ='violet'
            )
axes.set(xlabel='SPS_avg', ylabel='QTD.população')
```

O gráfico acima representa a média resultante da pesquisa COVIDiSTRESS em relação à escala SPS
